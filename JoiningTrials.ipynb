```python
import pandas as pd
import os

# Correct folder names
folders = ["results_clara", "results_erine", "results_zuriel", "results_joao"]
output_folder = "merged_trials"
os.makedirs(output_folder, exist_ok=True)

num_trials = 15

for i in range(1, num_trials + 1):
    dfs = []
    for folder in folders:
        file_path = os.path.join("02464_mini_project1/",folder, f"{folder}_{i}.csv")  # match exact filenames
        print(f"Processing file: {file_path}")
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            # add subject column
            df["subject"] = folder.split("_")[1]
            dfs.append(df)
        else:
            print(f"File not found: {file_path}")
    
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        combined_df.to_csv(os.path.join(output_folder, f"trial_{i}.csv"), index=False)
        print(f"Created merged CSV for trial {i}")

```
